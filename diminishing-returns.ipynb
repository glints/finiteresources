```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
resources = []
years = []

i = 1000
# 5% consumption a year
iteration = 0
while i > 0.5:
    i = i - ((i / 100) * 5)
    resources.append(i)
    years.append(iteration)
    iteration += 1
    
#t = np.c_[resources,years]
#t = np.flip(np.column_stack((years, resources)))
#pprint(years)
#t = np.array([years,resources],dtype = 'int')
#pprint(t)

# red dashes, blue squares and green triangles
#plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
#plt.plot(t)


#x,y
plt.plot(years,resources)
plt.title('5% annual consumption')
plt.ylabel('finite resources')
plt.xlabel('years')
plt.show()
```
