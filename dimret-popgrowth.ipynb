```python
import numpy as np
import matplotlib.pyplot as plt
from pprint import pprint
```

```python
resources = []
years = []
population = []
#bln
# https://www.infoplease.com/world/population-statistics/total-population-world-decade-1950-2050
populationstart = 2.2
# populationgrowth from 50 to 70 was roughly 20%, it then starts declining
populationgrowth = 1.5
consumptionpercentage = 2


finiteresources = 1000
# population * % consumption a year
year = 1950
#while finiteresources > 0.0000000001:
while finiteresources > 0.001:
    # lets take 1950 as a starting point for population == 2.2 bil
    # population grows 1.1% every year (2018)
    # https://data.worldbank.org/indicator/SP.POP.GROW?end=2018&start=1960&view=chart
    #after 20 years population growth starts declining about 2% a year
    if year > 1970:
        populationgrowth = populationgrowth - ((populationgrowth / 100) * 1.2)
        
    populationstart = populationstart + ((populationstart / 100) * populationgrowth)
    
    #i = i - (populationstart * ((i / 100) * consumptionpercentage))
    finiteresources = finiteresources - ((finiteresources / 100) * consumptionpercentage)
    resources.append(finiteresources)
    years.append(year)
    population.append(populationstart)
    year += 1
    
t = years
data1 = resources
data2 = population

fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (years)')
ax1.set_ylabel('finite resources', color=color)
#ax1.set_yscale('log')
ax1.plot(t, data1, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis
#ax2.set_yscale('log')
color = 'tab:blue'
ax2.set_ylabel('population (bln)', color=color)  # we already handled the x-label with ax1
ax2.plot(t, data2, color=color)
ax2.tick_params(axis='y', labelcolor=color)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
plt.title('2% annual consumption finite resources vs 1.5% demographic growth')
#plt.grid(True)
plt.show()
```
